```python
#C5. Будем считать, что выбросами считаются наблюдения, которые отличаются от среднего арифметического более, чем на 1.5 стандартных отклонения. Определите, сколько наблюдений считаются выбросами по признаку Price. Ответ округлите до двух знаков.

avg = df['Price'].mean()
std = df['Price'].std()

left = avg - 1.5 * std #1.5 заменить на значение стандартного отклонения
right = avg + 1.5 * std
result = ((df['Price'] < left) | (df['Price'] > right)).sum()
#тут в формуле надо внимательно смотреть на знаки, потому что в df[(df['Price'] >= left) & (df['Price'] >= right)].shape[0] другие диапазоны

result.round(2)
```

```python
#ОБЩЕЕ

import statsmodels.api as sm

Y = df['название_переменной_на_что_влияет']  # на что влияет
X = sm.add_constant(df['название_переменной_что_влияет'])  # то что влияет

model = sm.OLS(Y, X).fit()
Y_pr = model.predict(X)  # предсказанные значения

# Среднеквадратичная ошибка (MSE)
MSE = ((Y - Y_pr) ** 2).mean()
MSE_rounded = round(MSE, 2)  # округление до двух знаков

# Для средней абсолютной ошибки (MAE)
MAE = (abs(Y - Y_pr)).mean()
MAE_rounded = round(MAE, 2)  # округление до двух знаков

# Для коэффициента детерминации (R2)
R2 = model.rsquared
R2_rounded = round(R2, 2)  # округление до двух знаков

# Печатаем результаты
print("MSE:", MSE_rounded)
print("MAE:", MAE_rounded)
print("R2:", R2_rounded)
```
